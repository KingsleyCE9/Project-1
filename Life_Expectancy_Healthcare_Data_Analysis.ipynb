STEEP:8. Life Expectancy Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Life expectancy "], kde=True)
plt.title("Distribution of Life Expectancy")
plt.show()
```

STEEP:9.   Correlation Analysis





In this section, I looked at the relationships between healthcare indicators and life expectancy to understand how the variables in the dataset are connected to each other.

```python
corr = df.corr(numeric_only=True)

plt.figure(figsize=(10,6))
sns.heatmap(corr)
plt.title("Correlation Heatmap")
plt.show()
```
