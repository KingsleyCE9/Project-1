Load dataset:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load CSV
df = pd.read_csv('diabetes.csv')
df.head()
```

Check basic info:

```python
df.info()
df.describe()
df.isnull().sum()
```

Save figures directory

```python
import os
os.makedirs('figures', exist_ok=True)
```

Visualizations

A. Outcome distribution (bar chart)

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Outcome', data=df)
plt.xlabel('Outcome (0 = No Diabetes, 1 = Diabetes)')
plt.title('Distribution of Diabetes Outcome')
plt.tight_layout()
plt.savefig('figures/outcome_distribution.png', bbox_inches='tight')
plt.show()
```

B. Histograms + KDE for numeric features (combined)

```python
num_cols = ['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']
plt.figure(figsize=(16,12))
for i, col in enumerate(num_cols, 1):
    plt.subplot(4,2,i)
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'{col} distribution')
plt.tight_layout()
plt.savefig('figures/histograms_kde.png', bbox_inches='tight')
plt.show()
```

C. Boxplots for outlier detection (one per variable)

```python
plt.figure(figsize=(16,12))
for i, col in enumerate(num_cols, 1):
    plt.subplot(4,2,i)
    sns.boxplot(x=df[col])
    plt.title(f'{col} boxplot')
plt.tight_layout()
plt.savefig('figures/boxplots.png', bbox_inches='tight')
plt.show()
```

D. Correlation heatmap (features and Outcome)

```python
plt.figure(figsize=(10,8))
corr = df.corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.tight_layout()
plt.savefig('figures/correlation_heatmap.png', bbox_inches='tight')
plt.show()
```

Pairplot colored by Outcome (slower on big data but fine here)

```python
# pairplot can be slow, limit to a subset of columns for clarity
cols = ['Glucose','BMI','Age','Insulin','BloodPressure','Outcome']
sns.pairplot(df[cols], hue='Outcome', corner=True, plot_kws={'s':20})
plt.savefig('figures/pairplot.png', bbox_inches='tight')
plt.show()
```

Violin or boxplot of key features by Outcome

```python
features = ['Glucose','BMI','Age','Insulin']
plt.figure(figsize=(12,8))
for i, f in enumerate(features,1):
    plt.subplot(2,2,i)
    sns.violinplot(x='Outcome', y=f, data=df, inner='quartile')
    plt.title(f'{f} distribution by Outcome')
plt.tight_layout()
plt.savefig('figures/violin_by_outcome.png', bbox_inches='tight')
plt.show()
```

Rolling trend or grouped summary

```python
df['AgeGroup'] = pd.cut(df['Age'], bins=[20,30,40,50,60,100], labels=['21-30','31-40','41-50','51-60','61+'])
age_group = df.groupby('AgeGroup')['Outcome'].mean().reset_index()
plt.figure(figsize=(8,4))
sns.barplot(x='AgeGroup', y='Outcome', data=age_group)
plt.ylabel('Proportion with Diabetes')
plt.title('Diabetes Prevalence by Age Group')
plt.tight_layout()
plt.savefig('figures/diabetes_by_agegroup.png', bbox_inches='tight')
plt.show()
```
