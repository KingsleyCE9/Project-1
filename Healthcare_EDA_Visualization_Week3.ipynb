```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots look nice
sns.set()
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("diabetic_data.csv")

# Show first 5 rows
df.head()
```

```python
# Summary statistics
df.describe()
```

```python
# Count missing values
df.isnull().sum()
```

A â€” Age distribution (or age bucket counts)

```python
plt.figure(figsize=(10,5))
sns.countplot(x="age", data=df, palette="viridis")
plt.title("Distribution of Patients by Age Group")
plt.xticks(rotation=45)
plt.show()
```

B â€” Gender counts

```python
sns.countplot(x="gender", data=df)
plt.title("Patients by Gender")
plt.show()
```

C â€” Readmission counts

```python
import os

# Create "figures" folder if it doesn't exist
os.makedirs("figures", exist_ok=True)
```

```python
plt.figure(figsize=(6,4))
if 'readmitted' in df.columns:
    sns.countplot(x='readmitted', data=df, order=df['readmitted'].cat.categories if hasattr(df['readmitted'], 'cat') else None)
    plt.title('Readmission categories')
plt.savefig('figures/readmit_counts.png', bbox_inches='tight')
plt.show()
```

D â€” Time in hospital distribution

```python
plt.figure(figsize=(8,4))
if 'time_in_hospital' in df.columns:
    sns.histplot(df['time_in_hospital'].dropna(), kde=True)
    plt.title('Time in hospital distribution')
    plt.savefig('figures/time_in_hospital.png', bbox_inches='tight')
    plt.show()
```

E â€” Boxplot: time_in_hospital by readmit flag

```python
plt.figure(figsize=(8,5))
if 'readmit_30d' in df.columns and 'time_in_hospital' in df.columns:
    sns.boxplot(x='readmit_30d', y='time_in_hospital', data=df)
    plt.title('Time in hospital by 30-day readmit flag')
    plt.savefig('figures/los_by_readmit.png', bbox_inches='tight')
    plt.show()
```

F â€” Correlation heatmap (numeric only)

```python
plt.figure(figsize=(12,10))
num_df = df.select_dtypes(include=['int64','float64'])
sns.heatmap(num_df.corr(), annot=True, fmt='.2f', cmap='vlag')
plt.title('Numeric correlation heatmap')
plt.savefig('figures/corr_heatmap.png', bbox_inches='tight')
plt.show()
```
